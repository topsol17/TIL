```python
def second_index(text, key):
    count = 0
    index = None
    
    for i in range(len(text)):
        if text[i] == key:
            index = i
            count += 1

        if count > 1:
            break
        
    return index
```

```python
assert second_index('sims', 's') == 3
assert second_index('find the river', 'e') == 12
assert second_index('hi', ' ') is None
```
